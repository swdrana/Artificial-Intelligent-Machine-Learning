```python
import numpy as np
import pandas as pd

np.random.seed(24)

num_employees = 50  # you can increase this if needed

# Employee data
emp_ids = np.arange(101, 101 + num_employees)
emp_ages = np.random.randint(20, 55, num_employees)
emp_income = np.random.randint(35000, 95000, num_employees)

# Fixed set of roles (repetition allowed)
roles_list = [
    "HR Manager",
    "Software Developer",
    "Accountant",
    "Marketing Executive",
    "Sales Executive",
    "Backend Developer",
    "HR Executive",
    "Financial Analyst",
    "Digital Marketer",
    "Operations Manager"
]

emp_roles = np.random.choice(roles_list, num_employees)

# Create DataFrame
data = pd.DataFrame({
    "emp_id": emp_ids,
    "emp_age": emp_ages,
    "emp_salary": emp_income,
    "role": emp_roles
})

# Save dataset
data.to_csv("staff_data.csv", index=False)
```

Q4.
Create a new NumPy array where each employee’s age is adjusted based on:

Add 2 years if age is greater than or equal to the average age

Print:

Q5.
Using NumPy:
Assume all salaries of Tech roles increase by 8%
Calculate the new total salary expense
Find the overall increase in expense after the raise

Print both old and new totals along with the difference.

```python
tech_roles = ['Software Developer', 'Backend Developer']
is_tech = np.isin(emp_roles, tech_roles)

new_salaries = emp_income.copy()
new_salaries[is_tech] = new_salaries[is_tech] * 1.08

old_total = np.sum(emp_income)
new_total = np.sum(new_salaries)
difference = new_total - old_total

print(old_total)
print(new_total)
print(difference)
```

Q10.
Using .loc:
Increase salary by 12% for "Software Developer" and "Backend Developer"
Increase salary by 5% for all other roles
Then:
Print the dataframe that includes role , previous_salary , new_salary , increase_of_salary

```python
df_salary = df.copy()
df_salary['emp_salary'] = df_salary['emp_salary'].astype(float)
df_salary['previous_salary'] = df_salary['emp_salary']

tech_roles = ['Software Developer', 'Backend Developer']

df_salary.loc[df_salary['role'].isin(tech_roles), 'emp_salary'] = (
    df_salary.loc[df_salary['role'].isin(tech_roles), 'emp_salary'] * 1.12
).round(2)

df_salary.loc[~df_salary['role'].isin(tech_roles), 'emp_salary'] = (
    df_salary.loc[~df_salary['role'].isin(tech_roles), 'emp_salary'] * 1.05
).round(2)

df_salary['new_salary'] = df_salary['emp_salary']
df_salary['increase_of_salary'] = (df_salary['new_salary'] - df_salary['previous_salary']).round(2)

result = df_salary[['role', 'previous_salary', 'new_salary', 'increase_of_salary']]
print(result.to_string(index=False))
```
