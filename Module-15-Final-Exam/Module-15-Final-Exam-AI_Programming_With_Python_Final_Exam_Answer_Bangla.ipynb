## 🔰 0) Starter Code — সবার আগে এটা Run করো

````
import numpy as np          # সংখ্যার array নিয়ে কাজ করার library
import pandas as pd         # টেবিল (DataFrame) নিয়ে কাজ করার library
import matplotlib.pyplot as plt  # চার্ট আঁকার library
import seaborn as sns       # আরও সুন্দর চার্ট আঁকার library

np.random.seed(42)          # seed দিলে প্রতিবার একই random সংখ্যা আসে — result reproducible হয়
num_students = 50           # মোট student সংখ্যা

student_ids = np.arange(201, 201 + num_students)              # 201 থেকে 250 পর্যন্ত ID তৈরি
student_ages = np.random.randint(15, 22, num_students)        # 15-21 বছরের মধ্যে random বয়স
student_scores = np.random.randint(40, 100, num_students)     # 40-99 এর মধ্যে random score
student_attendance = np.random.randint(60, 100, num_students) # 60-99% এর মধ্যে random উপস্থিতি

departments_list = ["Science","Commerce","Arts","Computer Science",
                    "Mathematics","English","Physics","Chemistry","Biology","Economics"]

student_departments = np.random.choice(departments_list, num_students) # random department বাছাই

# প্রতিটি score দেখে grade ঠিক করার condition
grade_conditions = [
    student_scores >= 80,                                    # 80+ হলে A
    (student_scores >= 65) & (student_scores < 80),         # 65-79 হলে B
    (student_scores >= 50) & (student_scores < 65),         # 50-64 হলে C
    student_scores < 50                                      # 50 এর নিচে হলে F
]
grade_values = ["A", "B", "C", "F"]
student_grades = np.select(grade_conditions, grade_values, default="F")  # condition মিলিয়ে grade দাও

# সব array দিয়ে একটা টেবিল (DataFrame) তৈরি করো
data = pd.DataFrame({
    "student_id": student_ids,
    "student_age": student_ages,
    "score": student_scores,
    "attendance": student_attendance,
    "department": student_departments,
    "grade": student_grades
})

data.to_csv("school_data.csv", index=False)  # CSV হিসেবে সেভ করো (index column ছাড়া)
print("✅ Dataset তৈরি ও save সম্পন্ন!")
````

## ✅ Question 01 — NumPy Array তৈরি ও Properties [10 Marks]
**কী শিখব:** NumPy array-এর dtype, ndim, shape, size কী মানে।
| Property | মানে |
|---|---|
| `.dtype` | data-র ধরন (যেমন int64) |
| `.ndim` | কতটি মাত্রা (1D হলে 1) |
| `.shape` | প্রতি মাত্রায় কয়টি element (50,) |
| `.size` | মোট element সংখ্যা |

````
# Answer 01
score_array = np.array(data['score'])          # 'score' column → NumPy array
attendance_array = np.array(data['attendance']) # 'attendance' column → NumPy array

print("=== Score Array ===")
print(score_array)                             # পুরো array print করো
print("Data Type      :", score_array.dtype)   # কী ধরনের সংখ্যা (int32/int64)
print("Dimensions     :", score_array.ndim)   # মাত্রা সংখ্যা → 1 (একটি মাত্রা)
print("Shape          :", score_array.shape)  # আকার → (50,) মানে ৫০টি element
print("Total Elements :", score_array.size)   # মোট element → 50

print()
print("=== Attendance Array ===")
print(attendance_array)
print("Data Type      :", attendance_array.dtype)
print("Dimensions     :", attendance_array.ndim)
print("Shape          :", attendance_array.shape)
print("Total Elements :", attendance_array.size)
````

## ✅ Question 02 — Statistics ও Filtering [10 Marks]
**Boolean Indexing:** `arr[arr > value]` → condition সত্য হলে সেই element রাখো, মিথ্যা হলে বাদ দাও।

````
# Answer 02
score_array = np.array(data['score'])
attendance_array = np.array(data['attendance'])

max_score      = np.max(score_array)       # সর্বোচ্চ score
min_score      = np.min(score_array)       # সর্বনিম্ন score
avg_score      = np.mean(score_array)      # গড় score
avg_attendance = np.mean(attendance_array) # গড় উপস্থিতি

print("Maximum Score      :", max_score)
print("Minimum Score      :", min_score)
print("Average Score      :", avg_score)
print("Average Attendance :", avg_attendance)

# Boolean Indexing — গড়ের বেশি score যাদের তাদের বের করো
# score_array > avg_score  →  [True, False, True, ...]  এই ধরনের array তৈরি হয়
# score_array[সেই array]  →  শুধু True যেগুলো সেগুলো নেওয়া হয়
above_avg_scores = score_array[score_array > avg_score]

print("\nStudents with Score Greater than Average Score:")
print(above_avg_scores)
````

## ✅ Question 03 — Pandas CSV Load ও Filtering [10 Marks]
- `read_csv()` → CSV ফাইল পড়ো  
- `head(n)` → প্রথম n row  
- `info()` → column নাম, dtype, null count  
- `describe()` → গড়, min, max, std  
- `&` → দুই condition একসাথে (Python-এর `and` না)  
- `tail(n)` → শেষ n row

````
# Answer 03
df = pd.read_csv("school_data.csv")  # CSV ফাইল পড়ে DataFrame তৈরি করো

print("=== First 3 Rows ===")
print(df.head(3))    # প্রথম ৩টি row দেখাও

print("\n=== Dataset Information ===")
df.info()            # column নাম, data type, null কতটা — সব দেখাও

print("\n=== Summary Statistics ===")
print(df.describe()) # গড়, min, max, std — numerical column-এর সারসংক্ষেপ

# শুধু ৪টি column নাও (list দিয়ে select করো)
filtered_df = df[['student_id', 'student_age', 'department', 'grade']]

# দুটো condition একসাথে: age >= 16 AND age <= 20
# প্রতিটি condition আলাদা () এ রাখতে হয়, & দিয়ে জুড়তে হয়
age_filtered = filtered_df[
    (filtered_df['student_age'] >= 16) & (filtered_df['student_age'] <= 20)
]

sorted_df = age_filtered.sort_values(by='student_id')  # student_id অনুযায়ী ছোট থেকে বড় সাজাও

print("\n=== Last 3 Rows (Age 16-20, Sorted by Student ID) ===")
print(sorted_df.tail(3))  # শেষ ৩টি row দেখাও
````

## ✅ Question 04 — `.loc` দিয়ে Score Update [10 Marks]
```
df.loc[কোন row, কোন column] = নতুন মান
```
- `isin([...])` → list-এর যেকোনোটা মিললে True  
- `~` → NOT (বিপরীত)  
- `.clip(upper=100)` → ১০০ এর বেশি হলে জোর করে ১০০ করো

````
# Answer 04
df = pd.read_csv("school_data.csv")

df['previous_score'] = df['score']  # আগের score সংরক্ষণ করো (পরে তুলনার জন্য)

# isin() → list-এর যেকোনোটা মিললে True — এখানে CS বা Math হলে True
df.loc[df['department'].isin(['Computer Science', 'Mathematics']), 'score'] = (
    df.loc[df['department'].isin(['Computer Science', 'Mathematics']), 'score'] * 1.10  # ১০% বাড়াও
)

# ~ মানে NOT — CS বা Math নয় এমন department
df.loc[~df['department'].isin(['Computer Science', 'Mathematics']), 'score'] = (
    df.loc[~df['department'].isin(['Computer Science', 'Mathematics']), 'score'] * 1.05  # ৫% বাড়াও
)

df['score'] = df['score'].clip(upper=100)          # ১০০ এর বেশি হলে ১০০ করো
df['new_score'] = df['score'].round(2)             # ২ দশমিক পর্যন্ত রাখো
df['score_increase'] = (df['new_score'] - df['previous_score']).round(2)  # বৃদ্ধি কতটুকু

result = df[['student_id', 'department', 'previous_score', 'new_score', 'score_increase']]
print(result.to_string(index=False))  # সুন্দরভাবে print করো (index ছাড়া)
````

## ✅ Question 05 — Bar Chart [10 Marks]
`groupby('department')['score'].mean()` → department অনুযায়ী ভাগ করো, প্রতিটির গড় বের করো

````
# Answer 05
df = pd.read_csv("school_data.csv")

# department অনুযায়ী group করো, প্রতিটি group-এর score-এর গড় বের করো
avg_score_by_dept = df.groupby('department')['score'].mean()

plt.figure(figsize=(12, 6))           # chart-এর size (প্রস্থ=12, উচ্চতা=6)
plt.bar(
    avg_score_by_dept.index,          # X-axis: department নাম
    avg_score_by_dept.values,         # Y-axis: গড় score
    color='steelblue',                # bar-এর রঙ
    edgecolor='black'                 # bar-এর বর্ডার রঙ
)
plt.title("Average Score by Department")  # chart-এর title
plt.xlabel("Department")                  # X-axis label
plt.ylabel("Average Score")               # Y-axis label
plt.xticks(rotation=45, ha='right')       # label ৪৫° কোণে দেখাও (overlap এড়াতে)
plt.tight_layout()                        # সব কিছু ফিট করো
plt.show()                                # chart দেখাও
````

## ✅ Question 06 — Histogram [10 Marks]
**Histogram কী?** কতজন student কোন score range-এ আছে দেখায়।  
`bins=10` → score range কে ১০টি ভাগে ভাগ করো

````
# Answer 06
df = pd.read_csv("school_data.csv")

plt.figure(figsize=(10, 6))
plt.hist(
    df['score'],              # score column-এর data নাও
    bins=10,                  # ১০টি ভাগে ভাগ করো (40-46, 46-52, ...)
    color='cornflowerblue',   # bar-এর রঙ
    edgecolor='black'         # bar-এর বর্ডার
)
plt.title("Distribution of Student Scores")  # title
plt.xlabel("Score")                          # X-axis: score range
plt.ylabel("Number of Students")             # Y-axis: কতজন student
plt.tight_layout()
plt.show()
````

## ✅ Question 07 — Scatter Plot [10 Marks]
**Scatter Plot কী?** দুটো জিনিসের সম্পর্ক দেখায়। প্রতিটি student = একটি বিন্দু।  
`alpha=0.7` → বিন্দু ৭০% অস্বচ্ছ (একটু transparent)

````
# Answer 07
df = pd.read_csv("school_data.csv")

plt.figure(figsize=(10, 6))
plt.scatter(
    df['attendance'],    # X-axis: উপস্থিতির হার (%)
    df['score'],         # Y-axis: score
    color='darkorange',  # বিন্দুর রঙ
    edgecolors='black',  # বিন্দুর বর্ডার
    alpha=0.7            # স্বচ্ছতা (0=পুরো স্বচ্ছ, 1=অস্বচ্ছ)
)
plt.title("Attendance vs Score")
plt.xlabel("Attendance (%)")   # X-axis label
plt.ylabel("Score")            # Y-axis label
plt.tight_layout()
plt.show()
````

## ✅ Question 08 — Seaborn Boxplot [10 Marks]
**Boxplot কী?**
- বক্সের মাঝের রেখা = **median** (মাঝামাঝি মান)
- বক্সের উপর-নিচ সীমা = ৭৫% ও ২৫% মান
- লাইনের শেষ = সর্বোচ্চ/সর্বনিম্ন

````
# Answer 08
df = pd.read_csv("school_data.csv")

plt.figure(figsize=(10, 6))
sns.boxplot(
    x='grade',                    # X-axis: grade (A, B, C, F)
    y='score',                    # Y-axis: score
    data=df,                      # কোন DataFrame থেকে data নেবে
    palette='Set2',               # রঙের সেট (Set1, Set2, tab10 ইত্যাদি)
    order=['A', 'B', 'C', 'F']   # grade-এর ক্রম নির্ধারণ করো
)
plt.title("Score Distribution by Grade")
plt.xlabel("Grade")
plt.ylabel("Score")
plt.tight_layout()
plt.show()
````

## ✅ Question 09 — Seaborn Countplot [10 Marks]
**Countplot কী?** প্রতিটি category-তে কতটি data আছে নিজে গণনা করে bar আঁকে।

````
# Answer 09
df = pd.read_csv("school_data.csv")

plt.figure(figsize=(12, 6))
sns.countplot(
    x='department',   # X-axis: department নাম
    data=df,          # DataFrame
    palette='viridis' # রঙের থিম (viridis, magma, rocket, tab10 ইত্যাদি)
)
plt.title("Number of Students per Department")
plt.xlabel("Department")
plt.ylabel("Number of Students")          # Y-axis: Seaborn নিজেই count করে
plt.xticks(rotation=45, ha='right')       # label ৪৫° কোণে (overlap এড়াতে)
plt.tight_layout()
plt.show()
````

## ✅ Question 10 — Seaborn Scatter রঙ দিয়ে [10 Marks]
`hue='department'` → এটাই মূল বিষয়! `hue` মানে কোন column-এর উপর ভিত্তি করে রঙ দেবে।

````
# Answer 10
df = pd.read_csv("school_data.csv")

plt.figure(figsize=(12, 7))
sns.scatterplot(
    x='attendance',      # X-axis: উপস্থিতির হার
    y='score',           # Y-axis: score
    hue='department',    # department অনুযায়ী আলাদা রঙ দাও ← সবচেয়ে গুরুত্বপূর্ণ অংশ
    data=df,
    palette='tab10',     # ১০টি আলাদা রঙের থিম (১০টি department-এর জন্য ভালো)
    s=100,               # বিন্দুর আকার (size)
    edgecolor='black',   # বিন্দুর বর্ডার
    alpha=0.8            # সামান্য স্বচ্ছতা
)
plt.title("Attendance vs Score by Department")
plt.xlabel("Attendance (%)")
plt.ylabel("Score")
# legend কে chart-এর বাইরে ডান দিকে রাখো
plt.legend(title='Department', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()  # legend সহ সব ফিট করো
plt.show()
````

## 🧠 Quick Reference

| কাজ | কোড |
|---|---|
| NumPy array | `np.array(data['col'])` |
| max/min/mean | `np.max()`, `np.min()`, `np.mean()` |
| Boolean filter | `arr[arr > value]` |
| CSV পড়া | `pd.read_csv("file.csv")` |
| Column select | `df[['col1','col2']]` |
| Row filter | `df[(cond1) & (cond2)]` |
| Group করা | `df.groupby('col')['score'].mean()` |
| Row/col update | `df.loc[condition, 'col'] = value` |
| সীমা নির্ধারণ | `df['col'].clip(upper=100)` |
| Bar chart | `plt.bar(x, y)` |
| Histogram | `plt.hist(data, bins=10)` |
| Scatter | `plt.scatter(x, y)` |
| Boxplot | `sns.boxplot(x=, y=, data=)` |
| Count chart | `sns.countplot(x=, data=)` |
| রঙিন scatter | `sns.scatterplot(hue='col')` |
